```python
import numpy as np
import random
import networkx as nx
from IPython.display import Image
import matplotlib.pyplot as plt
```

```python
G = nx.grid_graph([4, 4])
```

```python
pos = nx.spring_layout(G)
# Plot the graph
nx.draw(G, cmap = plt.get_cmap('rainbow'), with_labels=True, pos=pos)
```
